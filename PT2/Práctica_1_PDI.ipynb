###3. Interacción con el usuario
Utilizaremos botones scroll para que el usuario seleccione el valor de R, G Y B, siendo:

*   R: coeficiente de la componente RED
*   G: coeficiente de la componente GREEN
*   B: coeficiente de la componente BLUE

```python
R = 0.7 #@param {type:"slider", min:0, max:1, step:0.1}
G = 0.4 #@param {type:"slider", min:0, max:1, step:0.1}
B = 0.6 #@param {type:"slider", min:0, max:1, step:0.1}
```

###4. Desarrollo
La matriz (matrixRGB) almacena los coeficientes del espacio RGB (comprendidos entre 0 y 1) elegido por el usuario.


```python
matrixRGB=np.array([[R,0,0],[0,G,0],[0,0,B]])
image=imageio.imread('/content/drive/MyDrive/Colab Notebooks/loro.jpg')#Lectura de la imagen
```

Se puede visualizar la imagen en un arreglo de matrices

```python
print(image)
```

Los comandos TYPE e IMAGEN.SHAPE permite visualizar el tipo, tamaño y canales de la imagen. Se normaliza los valores RGB de la imagen dividiendo la matriz por 255. Normalizado los valores RGB, multiplicaremos por los coeficientes seleccionados por el usuario, para obtener la nueva imagen con alteración en los canales RGB. Utilizando el comando imageio.imwrite se procede a guardar la nueva imagen.

```python
print('This image is', type(image))
print('with dimensions', image.shape)

plt.figure()
plt.title('Original Image')
plt.imshow(image)
plt.colorbar()
plt.grid(False)
plt.show()

image=image/255 #Normalización de valores RGB
plt.figure()
plt.title('Normalized RGB image')
plt.imshow(image)
plt.colorbar()
plt.grid(False)
plt.show()

image=np.dot(image,matrixRGB)
plt.figure()
plt.title('RGB Coefficiente')
plt.imshow(image)
plt.colorbar()
plt.grid(False)
plt.show()

imageio.imwrite('/content/drive/MyDrive/Colab Notebooks/loro.2.jpg', (image*255).astype(np.uint8))
imagensave=imageio.imread('/content/drive/MyDrive/Colab Notebooks/loro.2.jpg')
plt.figure()
plt.title('Save Image')
plt.imshow(image)
plt.colorbar()
plt.grid(False)
plt.show()
```
